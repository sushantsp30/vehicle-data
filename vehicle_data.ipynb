```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('cars.csv')
```

```python
df.describe()

```

```python
df.info()
```

```python
df['weightlbs'] = pd.to_numeric(df['weightlbs'], errors='coerce')
```

```python
df['cubicinches'] = df['cubicinches'].astype(int)
```

```python
df.info()
```

```python
df['weightlbs'].fillna(df['weightlbs'].mean(), inplace=True)
```

```python
df['cubicinches'].fillna(df['cubicinches'].mean(), inplace=True)
```

1.Can you find that there is improvement in mpg over the years ?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='mpg', data=df)
plt.title('MPG vs Year')
plt.xlabel('Year')
plt.ylabel('MPG')
plt.show()
```

Increasing years there is increase in  mpg

2.Does cubicinches of a vehicle affects the vehicles time to reach 60 Miles per hour?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='cubicinches', y='time-to-60', data=df)
plt.title('Cubic Inches vs Time to 60 MPH')
plt.xlabel('Cubic Inches')
plt.ylabel('Time to 60 MPH')
plt.show()
```

Increase in cubic inches the time taken to reach 60 decreases

3.Does a specific country prefer vehicles with more horse power ?

```python
df.columns = df.columns.str.strip()

plt.figure(figsize=(10, 6))
sns.barplot(x='brand', y='hp', data=df)
plt.title('Horsepower vs Brand')
plt.xlabel('Brand')
plt.ylabel('Horsepower')
plt.xticks(rotation=45)
plt.show()
```

US prefers vehicles with more horse power

4.How does the horse power affects the average of the vehicle ?

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='hp', y='mpg', data=df, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Horsepower vs MPG')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.show()
```

The dataset contains 261 vehicle records, and after removing outliers, 252 records remain, showing that vehicles with 8 cylinders make up approximately 29.1% of the total records.

5.How does the cubic inches affects the average of the vehicle ?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='mpg', y='cubicinches', data=df)
plt.title('Cubic Inches vs MPG')
plt.xlabel('MPG')
plt.ylabel('Cubic Inches')
plt.show()
```

Less the cubic inches more will be the mpg 

6.What is the effect of number of cylinders on the mpg ?

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='cylinders', y='mpg', data=df)
plt.title('Effect of Number of Cylinders on MPG')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.show()
```

The dataset shows that vehicles with 8 cylinders constitute approximately 29.1% of the total records, and there is a noticeable trend of increasing mpg over the years.

7.What relation between cubic inches , number of cylinders and horse power you can observe from the data ?

```python
sns.pairplot(df, vars=['cubicinches', 'cylinders', 'hp'], kind='reg', diag_kind='kde')
plt.suptitle('Relationship between Cubic Inches, Number of Cylinders, and Horsepower', y=1.02)
plt.show()
```

The dataset shows that vehicles with 8 cylinders constitute approximately 29.1% of the total records, and there is a noticeable trend of increasing mpg over the years.

8.How much proportion of the total records is occupied by vehicles with 8 cylinders ?

```python
proportion_8_cylinders = (df['cylinders'] == 8).mean()

proportion_other_cylinders = 1 - proportion_8_cylinders

labels = ['8 Cylinders', 'Other Cylinders']
sizes = [proportion_8_cylinders, proportion_other_cylinders]
colors = ['#ff9999','#66b3ff']
explode = (0.1, 0)  

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Proportion of Vehicles with 8 Cylinders')
plt.show()
```

29.1% of the total records is occupied by vehicles with 8 cylinders
